```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## Local Government Area Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average maths score 

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

* Calculate the percentage of students with a passing reading score (50 or greater)

* Calculate the percentage of students who passed maths **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Calculate the total number of schools
total_schools=school_data["school_name"].count()
total_schools

# Calculate the total number of students
total_students=student_data["student_name"].count()
total_students

# Calculate the total budget
total_budget=school_data["budget"].sum()
total_budget

# Calculate the average maths score
average_maths=student_data[("maths_score")].mean()
average_maths

# Calculate the average reading score
average_reading=student_data[("reading_score")].mean()
average_reading

# Calculate the percentage of students with a passing maths score (50 or greater)
perc_maths_pass=(school_data_complete["maths_score"]>=50).sum() / school_data_complete["maths_score"].count()* 100
perc_maths_pass

# Calculate the percentage of students with a passing reading score (50 or greater)
perc_reading_pass= (school_data_complete["reading_score"]>=50).sum() / school_data_complete["reading_score"].count()* 100
perc_reading_pass

# Calculate the percentage of students who passed maths and reading (% Overall Passing)

Overall_passing_rate= school_data_complete[(school_data_complete['maths_score'] >= 50) & (school_data_complete['reading_score'] >= 50)]['Student ID'].count()/total_students*100
Overall_passing_rate
```

```python
# Create a dataframe to hold the above results
Df_Summary = pd.DataFrame({"Total Schools":[total_schools], "Total Students":[total_students],"Total Budget":[total_budget],
                                "Average Math Score":[average_maths], "Average Reading Score":[average_reading], "% Passing Maths":[perc_maths_pass],
                                "% Passing Reading":[perc_reading_pass],"% Overall Passing Rate":[Overall_passing_rate]})
Df_Summary
```

```python
#  give the displayed data cleaner formatting

Df_Summary["Total Students"]=Df_Summary["Total Students"].map("{:,}".format)
Df_Summary["Total Budget"]=Df_Summary["Total Budget"].map("${:,}".format)

Df_Summary
```

## School Summary

```python
#  let's check school name and type
school_types=school_data.set_index(["school_name"])["type"]
school_types.head()
```

* Create an overview table that summarises key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed maths **and** reading.)
  
* Create a dataframe to hold the above results

```python
# Total Students in each school
students_per_school=school_data_complete["school_name"].value_counts()
students_per_school.head()
```

```python
#Total School Budget
school_budget=school_data_complete.groupby(["school_name"])["budget"].mean()
school_budget.head()
```

```python
#Per Student Budget
per_student_budget=school_budget/students_per_school
per_student_budget.head()
```

```python
#Average Maths Score per school

average_maths_score_perSchool=school_data_complete.groupby(["school_name"])["maths_score"].mean()
average_maths_score_perSchool.head()
```

```python
#Average Reading Score per school
average_reading_score_perSchool= school_data_complete.groupby(["school_name"])["reading_score"].mean()
average_reading_score_perSchool.head()
```

```python
#% Percentage of Passing Maths per school 
passing_maths_score=school_data_complete.loc[school_data_complete["maths_score"]>=50]
passing_maths_score

#assign the math score by each school
group_maths_school=passing_maths_score["school_name"].value_counts()
group_maths_school.head()

#get the final percentage result
percentage_of_maths=group_maths_school/students_per_school*100
percentage_of_maths
```

```python
#% Percentage of Passing Reading per school 

passing_reading_score=school_data_complete.loc[school_data_complete["reading_score"]>=50]
passing_reading_score

#assign the reading score by each school
by_reading_PerSchool= passing_reading_score["school_name"].value_counts()
by_reading_PerSchool.head()

#get the final percentage result
percentage_of_reading=by_reading_PerSchool/students_per_school*100
percentage_of_reading
```

```python
# % Overall Passing (The percentage of students that passed maths and reading.)

Overall_passing_rate = school_data_complete.loc[(school_data_complete['maths_score'] >= 50) & (school_data_complete['reading_score'] >= 50)]["school_name"].value_counts()/students_per_school*100
Overall_passing_rate
```

```python
school_summary_df = pd.DataFrame({"School Type":school_types, "Total Students":students_per_school, "Total School Budget":school_budget
                                ,"Per Student Budget":per_student_budget ,"Average Maths Score" :average_maths_score_perSchool,
                             "Average Reading Score":average_reading_score_perSchool,
                             "% Passing Maths":percentage_of_maths, "% Passing Reading":percentage_of_reading,
                            "% Overall Passing ":Overall_passing_rate})
school_summary_df
```

```python
#  give the displayed data cleaner formatting
school_summary_df["Total School Budget"]=school_summary_df["Total School Budget"].map("${:,.2f}".format)
school_summary_df["Per Student Budget"]=school_summary_df["Per Student Budget"].map("${:,.2f}".format)

school_summary_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# create the bins.
bins = [0, 585, 630, 645, 680]
group_names = ["<$585", "$585-630", "$630-645", "$645-680"]

# Create a table that breaks down school performances based on average Spending Ranges (Per Student)

school_summary_df["Spending ranges (per student)"]= pd.cut(per_student_budget, bins, labels=group_names)

# Calculate the values for the data table

Average_math_scores = school_summary_df.groupby(["Spending ranges (per student)"]).mean()["Average Maths Score"]
Average_reading_scores = school_summary_df.groupby(["Spending ranges (per student)"]).mean()["Average Reading Score"]
perc_passing_math = school_summary_df.groupby(["Spending ranges (per student)"]).mean()["% Passing Maths"]
perc_passing_reading = school_summary_df.groupby(["Spending ranges (per student)"]).mean()["% Passing Reading"]
overall = school_summary_df.groupby(["Spending ranges (per student)"]).mean()["% Overall Passing "]


 # Assemble into data frame
spending_summary = pd.DataFrame({"Average Maths Score" : Average_math_scores.map('{:.2f}'.format),
                                 "Average Reading Score": Average_reading_scores.map('{:.2f}'.format),
                                 "% Passing Math": perc_passing_math.map('{:.2f}'.format),
                                 "% Passing Reading": perc_passing_reading.map('{:.2f}'.format),
                                  "% Overall passing ": overall.map('{:.2f}'.format)})
                        

spending_summary

```
